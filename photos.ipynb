```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('himeji.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.title("Original")
plt.imshow(img)
plt.show()


gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.title("Gray")
plt.imshow(gray_img, cmap='gray')
plt.show()

print(gray_img)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_filter(img, kernel):
  img_height, img_width = img.shape
  filtered_img = np.zeros((img_height - 2, img_width - 2), dtype=np.float32)

  for y in range(1, img_height - 1):
    for x in range(1, img_width - 1):
      neighborhood = img[y - 1 : y + 2, x - 1 : x + 2]
      convolution_value = np.sum(neighborhood * kernel)
      filtered_img[y - 1 , x - 1] = convolution_value
  filtered_img = np.clip(filtered_img, 0, 255).astype(np.uint8)

  plt.figure(figsize=(10,4))

  plt.subplot(1,2,1)

  plt.title("Filtered")
  plt.imshow(filtered_img,cmap="gray")

```

```python
apply_filter(gray_img,np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
]))
```

```python
apply_filter(gray_img, np.array([
        [-1, -2, -1],
        [ 0,  0,  0],
        [ 1,  2,  1]
      ]))
```

```python
apply_filter(gray_img, np.array([
        [ 1, -2, -1],
        [ 2,  0,  2],
        [-1, -2,  1]
      ]))
```

```python
#Ogólne wykrywanie krawędzi
apply_filter(gray_img, np.array([
    [ 0, -1,  0],
    [-1,  5, -1],
    [ 0, -1,  0]
]))
```

```python
#Wykrywanie krawędzi skośnych
apply_filter(gray_img, np.array([
    [ 0,  1,  2],
    [-1,  0,  1],
    [-2, -1,  0]
]))
```
