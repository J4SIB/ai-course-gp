```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(4+5*X**2 + 2*Y**2))

surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)

np.random.seed(42)
x_scatter = np.random.uniform(-5, 5, 50)
y_scatter = np.random.uniform(-5, 5, 50)
z_scatter = np.sin(np.sqrt(x_scatter**2 + y_scatter**2)) + np.random.uniform(-0.1, 0.1, 50)
ax.scatter(x_scatter, y_scatter, z_scatter, color='red', s=50)

ax.view_init(elev=30, azim=45)

fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('Epic 3D Surface Plot with Scatter Points')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1, y1, score1 = map(float, input("Podaj współczynniki x, y i wynik pierwszego równania: ").split())
x2, y2, score2 = map(float, input("Podaj współczynniki x, y i wynik drugiego równania: ").split())

A = np.array([[x1, y1], [x2, y2]], dtype=float)
B = np.array([score1, score2], dtype=float)

D = np.linalg.det(A)
x_vals = np.linspace(x-10, x+10, 100)

if D != 0:
    A_x = A.copy()
    A_x[:, 0] = B
    D_x = np.linalg.det(A_x)

    A_y = A.copy()
    A_y[:, 1] = B
    D_y = np.linalg.det(A_y)

    x, y = D_x / D, D_y / D

    y_vals1 = (score1 - x1 * x_vals) / y1
    y_vals2 = (score2 - x2 * x_vals) / y2

    plt.plot(x_vals, y_vals1, label=f"{x1}x + {y1}y = {score1}")
    plt.plot(x_vals, y_vals2, label=f"{x2}x + {y2}y = {score2}")
    plt.scatter([x], [y], color='red', label=f"Rozwiązanie ({x:.2f}, {y:.2f})")


else:
    y_vals1 = (score1 - x1 * x_vals) / y1
    y_vals2 = (score2 - x2 * x_vals) / y2
    plt.plot(x_vals, y_vals1, label=f"{x1}x + {y1}y = {score1}")
    plt.plot(x_vals, y_vals2, label=f"{x2}x + {y2}y = {score2}")
    print("Układ jest nieoznaczony lub sprzeczny")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Rozwiązanie układu równań metodą Cramera")
plt.grid()
plt.show()
```
