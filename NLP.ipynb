```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt_tab')

text = "NLP to skrót od Natural Language Processing! Po polsku to Przetwarzanie Języka Naturalnego"
print(sent_tokenize(text))
print(word_tokenize(text))
```

```python
from nltk.stem import WordNetLemmatizer, SnowballStemmer
lemmatizer = WordNetLemmatizer()
stemmer = SnowballStemmer("english")
nltk.download('wordnet')


words = ["bring","brought","fall","fell","flick","flicked","broke","broken","think","thought"]
print("Word;Lemma;Stem")
for word in words:
  print(word, lemmatizer.lemmatize(word, pos='v'),stemmer.stem(word))
```

```python
import gensim
from gensim.downloader import load

model = load("glove-wiki-gigaword-100")
```

```python
print(model["dog"][:8])
print(model["cat"][:8])
print(model["hike"][:7])
```

```python
word = input('Wpisz słowo: ').lower()

if word in model:
  similar_words = model.most_similar(word, topn=10)
  for w, similarity in similar_words:
    print(f"{w}: {similarity}")
else:
  print("Słowa niema w słowniku :(")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def cos_sim(a,b):
  return cosine_similarity([a],[b])[0]

print(cos_sim(model["dog"],model["cat"]))
print(cos_sim(model["paper"],model["book"]))
print(cos_sim(model["count"],model["numbers"]))
```
